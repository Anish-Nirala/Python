```python
class CoffeeMachine:
    def __init__(self,listOfAvailableItems):
        self.listOfAvailableItems = listOfAvailableItems
        self.availableResource = {'Water':1000, 'Milk': 500, "Coffee": 250, 'Money': 0}
        self.requirement = {'Espresso':{'Water': 200, 'Milk': 100, 'Coffee': 50, 'Money': 2},
                            'Latte': {'Water': 400, 'Milk': 200, 'Coffee': 75, 'Money': 5},
                            'Cappuccino': {'Water': 600, 'Milk': 300, 'Coffee': 100, 'Money': 10}}
        self.itemsRequired = ["Water", "Milk", "Coffee"]
        
    def report(self):
        print("\n")
        for item, availability in self.availableResource.items():
            print(item, availability)
        print("\n")
    def isAvailableOrder(self, item):
        return item in self.listOfAvailableItems
    
    def procedure(self, order):
        isAvailable = True
        for item in self.itemsRequired:
            if(self.availableResource[item]<self.requirement[order][item]):
                isAvailable = False
                break
        if(isAvailable==False):
            print("\nSorry, there is not enough", item.lower()+".")
            return False
        else:
            quarters, dimes, nickles, pennies = map(int, input("\nInsert the coin(quarters, dimes, nickles, pennies): ").split()) 
            valueCoin = quarters*0.25 + dimes*0.10 + nickles*0.05 + pennies*0.01
            if(valueCoin<self.requirement[order]['Money']):
                print("\nSorry that's not enough money. Money refunded.")
            else:
                for item in self.itemsRequired:
                    self.availableResource[item]-=self.requirement[order][item]
                self.availableResource['Money']+=self.requirement[order]['Money']
                if(valueCoin>self.requirement[order]['Money']):
                    print("\nHere is ",str(round(valueCoin-self.requirement[order]['Money'], 2))+"$","dollars in change.")
                print("\nHere is your",order+".","Enjoy!.\n")
        return True
        
def main():
    isCoffeeMachineOn = False
    coffee  = CoffeeMachine(["Espresso", "Latte", "Cappuccino"])
    isCoffeeMachineOn = True
    isResourceAvailable = True
    while(isCoffeeMachineOn and isResourceAvailable):
        order = input("What would you like? ").capitalize()
        if(order=="Off"):
            isCoffeeMachineOn = False
            continue
        elif(order=="Report"):
            coffee.report()
        else:
            if(coffee.isAvailableOrder(order)==False):
                print("\nOops! ",order," which you want is not prepared in this machine. Please enter the order again. ThankYou.")
                continue
            else:
                if(coffee.procedure(order)==False):
                    isResourceAvailable = False
    else:
        if(isCoffeeMachineOn==False):
            print("\nCoffee machine is not available due to maintenance.")
        if(isResourceAvailable==False):
            print("\nCoffee machine is not available due to unavailability of resources.")
main()
```
